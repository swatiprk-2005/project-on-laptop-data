* Graphs representing relation between different attributes of laptop.


```python
sns.countplot(x='warranty',data=df)
plt.title("Warranty of laptop")
plt.xlabel("Years of warranty")
plt.ylabel("Number of laptops")
plt.show()
```

```python
counts=df["warranty"].value_counts()
plt.pie(counts,labels=counts.index,autopct="%1.1f%%")
plt.title("Brands of laptop available")
plt.show()

#pie chart is techincally useful but one category is larger then the other snall slices overlap and their labels become unreadable.so bar graph is more preferable.
```

```python
sns.countplot(x="warranty",hue="Ram",data=df)
plt.title("warranty according to Ram")
plt.show()
```

```python
plt.hist(df["price"],bins=10)
plt.title("price distribution")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
sns.histplot(df["price"],bins=9,kde=True)
plt.title("price distribution")
plt.show()
```

```python
plt.scatter(df["Ram"],df["price"])
plt.xlabel("RAM")
plt.ylabel("price")
plt.title("RAM vs PRICE")
plt.show()
```

```python
sns.scatterplot(x="Ram",y="price",data=df)
plt.title("ram vs price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="brand",y="price",data=df)
plt.xticks(rotation=90) #it rotates the xlabels
plt.title("brand vs price")
plt.show()
```

```python
sns.countplot(x="Ram",data=df)
plt.title("ram distribution")
plt.show()
```

```python
avg_price=df.groupby("brand")["price"].mean().sort_values()
avg_price.plot(kind="bar")
plt.title("Avg price")
plt.ylabel("price")
plt.show()
```

```python
sns.violinplot(x='brand', y='price', data=df)
plt.xticks(rotation=90)
plt.title("Price Distribution by Company")
plt.show()
```

```python
top10 = df.nlargest(10, 'price')

sns.barplot(x='brand', y='price', data=top10)
plt.title("Top 10 Expensive Laptops")
plt.show()
```
